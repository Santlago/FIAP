Pré Processamento

```python
dados.head()
```

```python
dados_cleaned = dados.iloc[:,6:10]
```

```python
dados_cleaned.dropna(inplace=True)
dados_cleaned
```

```python
continuous_columns = dados_cleaned.select_dtypes(include=['float64', 'int64']).columns

# Use iloc to select all rows for the continuous columns
continuous_data = dados_cleaned.loc[:, continuous_columns]
continuous_data.values
```

```python
dados_cleaned.iloc[218,:]
```

```python
# Separação dos dados de entrada e dados de saída
# X maiúsculo ----> Dados de entrada / Features / Variáveis independentes
# y minúsculo ----> Dados de saída / Target (alvo) / Variável dependente
# dados.iloc[linhas, colunas]
X = dados_cleaned.iloc[:, [0, 1, 3]] # Todas as linhas e as colunas de 6, 7, 9
X.head()
```

```python
X.values
```

```python
# Dados de saída
cl_energia_nuclear = dados_cleaned['Electricity from nuclear (TWh)']
y = cl_energia_nuclear.values
y
```

```python
# Separação em dados de treino e dados de teste
# https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html
# SEED
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=10)
X_train
```

```python
print(y_train.dtype)
```

```python
y_train = pd.Categorical(y_train).codes
```

Escolha do algoritmo

```python
# Instanciar o modelo
my_model = KNeighborsClassifier()
```

```python
# Treinar o modelo
my_model.fit(X_train, y_train)
```

```python
# Realizar as predições
y_predict = my_model.predict(X_test)
y_predict
```

```python
y_test
```

```python
# Check the type of y_test and y_predict
print(f"Type of y_test: {type(y_test)}")
print(f"Type of y_predict: {type(y_predict)}")

# Ensuring they're array of integers
y_test = y_test.astype(int)
y_predict = y_predict.astype(int)
```

```python
# Calcular a acurácia
acuracia = 100 * (accuracy_score(y_test, y_predict))
print(f'A acurácia do modelo baseado no algoritmo KNN é de {acuracia: .2f} %')
```

```python
# prompt: generate a dispersion graphic

import matplotlib.pyplot as plt

# Create a scatter plot of the data
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test)

# Add labels and title
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Dispersion Plot")

# Show the plot
plt.show()
```
